## Interpolation for Given Points

Given the points $(x_0, y_0) = (0, 0)$, $(x_1, y_1) = (1, 2)$, and $(x_2, y_2) = (2, 2)$, we aim to find the interpolating polynomial $P_2(x)$ that passes through these points using two different bases:

1. **Monomial Basis: $P(x) = p_2 x^2 + p_1 x + p_0$**
2. **Lagrange Basis: $P(x) = \sum_{k=0}^{2} y_k \ell_k(x)$**

### 1. Monomial Basis Solution

We seek a polynomial of the form:
$$
P_2(x) = p_2 x^2 + p_1 x + p_0
$$

Given the conditions:
$$
P(0) = 0 \implies p_0 = 0
$$
$$
P(1) = 2 \implies p_2 + p_1 = 2
$$
$$
P(2) = 2 \implies 4p_2 + 2p_1 = 2
$$

#### Solving for $p_1$ and $p_2$:

- From $p_2 + p_1 = 2$, we get:
  $$
  p_1 = 2 - p_2
  $$

- Substitute $p_1 = 2 - p_2$ into $4p_2 + 2p_1 = 2$:
  $$
  4p_2 + 2(2 - p_2) = 2 \implies 4p_2 + 4 - 2p_2 = 2 \implies 2p_2 + 4 = 2
  $$
  $$
  2p_2 = -2 \implies p_2 = -1
  $$

- Then, $p_1 = 2 - (-1) = 3$.

The interpolating polynomial in the monomial basis is:
$$
\boxed{P_2(x) = -x^2 + 3x}
$$

### 2. Lagrange Basis Solution

The Lagrange interpolating polynomial is:
$$
P(x) = \sum_{k=0}^{2} y_k \ell_k(x)
$$
where the Lagrange basis polynomials are:
$$
\ell_0(x) = \frac{(x - 1)(x - 2)}{(0 - 1)(0 - 2)} = \frac{(x - 1)(x - 2)}{2}
$$
$$
\ell_1(x) = \frac{(x - 0)(x - 2)}{(1 - 0)(1 - 2)} = -x(x - 2)
$$
$$
\ell_2(x) = \frac{(x - 0)(x - 1)}{(2 - 0)(2 - 1)} = \frac{x(x - 1)}{2}
$$

The interpolating polynomial becomes:
$$
P(x) = 0 \cdot \ell_0(x) + 2 \cdot (-x(x - 2)) + 2 \cdot \frac{x(x - 1)}{2}
$$
$$
P(x) = -2x(x - 2) + x(x - 1)
$$

#### Expanding:
$$
P(x) = -2(x^2 - 2x) + (x^2 - x) = -2x^2 + 4x + x^2 - x = -x^2 + 3x
$$

The interpolating polynomial in the Lagrange basis is:
$$
\boxed{P_2(x) = -x^2 + 3x}
$$

### Conclusion

Both approaches yield the same interpolating polynomial:
$$
\boxed{P_2(x) = -x^2 + 3x}
$$

Thus, the interpolating polynomial is independent of the choice of basis.


## Uniqueness of the Interpolating Polynomial

The interpolating polynomial is unique, which means that for a given set of points, there is exactly one polynomial that passes through all of them. To understand why this is the case, let's consider two interpolating polynomials, denoted as \( p_n(x) \) and \( q_n(x) \), both of which are of degree less than or equal to \( n \). 

### Assumption of Non-Uniqueness

Assume, for the sake of contradiction, that \( p_n(x) \) and \( q_n(x) \) are two distinct interpolating polynomials that fit the same \( n + 1 \) data points:
- \( (x_0, y_0) \)
- \( (x_1, y_1) \)
- \( \ldots \)
- \( (x_n, y_n) \)

Given that both polynomials fit the same points, we can say that:
$$
p_n(x_k) = y_k \quad \text{and} \quad q_n(x_k) = y_k \quad \text{for } k = 0, 1, \ldots, n.
$$

### Constructing the Difference

Now, consider the difference of the two polynomials:
$$
r_n(x) = p_n(x) - q_n(x).
$$

The polynomial \( r_n(x) \) will also be a polynomial, and it is of degree at most \( n \) (since both \( p_n \) and \( q_n \) are at most degree \( n \)). Furthermore, we have:
$$
r_n(x_k) = p_n(x_k) - q_n(x_k) = y_k - y_k = 0 \quad \text{for } k = 0, 1, \ldots, n.
$$

This means that \( r_n(x) \) has at least \( n + 1 \) roots (the points \( x_0, x_1, \ldots, x_n \)).

### Degree of Polynomials

According to the fundamental theorem of algebra, a non-zero polynomial of degree \( n \) can have at most \( n \) distinct roots. Since \( r_n(x) \) has \( n + 1 \) roots and is of degree at most \( n \), the only possibility is that \( r_n(x) \) is the zero polynomial:
$$
r_n(x) = 0 \quad \text{for all } x.
$$

### Conclusion

Therefore, we conclude that:
$$
p_n(x) - q_n(x) = 0 \implies p_n(x) = q_n(x) \quad \text{for all } x.
$$

This shows that the two interpolating polynomials \( p_n(x) \) and \( q_n(x) \) are indeed the same, contradicting our initial assumption that they were distinct. 

Thus, the uniqueness of the interpolating polynomial implies that it is the same regardless of the basis used, be it the monomial basis or the Lagrange basis. 

In summary, the polynomial \( P_n(x) \) that interpolates the data points is unique and independent of the method used to construct it. The resulting polynomial for the given points will always be:
$$
\boxed{P_2(x) = -x^2 + 3x}
$$

$$

### Step 1: Define the Lagrange Polynomials

The Lagrange polynomial $\ell_i(x)$ associated with the node $x_i$ is defined as:

$$
\ell_i(x) = \prod_{\substack{j=0 \\ j \neq i}}^{N} \frac{x - x_j}{x_i - x_j}
$$

This polynomial has the property that $\ell_i(x_j) = \delta_{ij}$ (the Kronecker delta), meaning that $\ell_i(x_j) = 1$ if $i = j$ and $\ell_i(x_j) = 0$ if $i \neq j$.

### Step 2: Consider the Constant Function

Let us consider the constant function $f(x) = c$ for any constant $c$. We can interpolate this constant function using the Lagrange polynomials:

$$
f(x) = \sum_{i=0}^{N} \ell_i(x) f(x_i)
$$

Since $f(x_i) = c$ for all $i$, this simplifies to:

$$
c = \sum_{i=0}^{N} \ell_i(x) c
$$

### Step 3: Factor Out the Constant

We can factor out the constant $c$ from the right-hand side:

$$
c = c \sum_{i=0}^{N} \ell_i(x)
$$

Assuming $c \neq 0$, we can divide both sides by $c$:

$$
1 = \sum_{i=0}^{N} \ell_i(x)
$$

### Conclusion

Thus, we have shown that 

$$
\sum_{i=0}^{N} \ell_i(x) = 1
$$

for all values of $x$, not just at the nodes $x_i$. This means that the Lagrange polynomials $\ell_0(x), \ell_1(x), \ldots, \ell_N(x)$ form a "Partition of Unity," as required.

To do this you must prove that

a & ~ & m = n \\ 
0 & ~ & m \neq n

where $a$ is a finite constant (also find this coefficient).  Note that you may have multiple non-zero coefficients $a$.  Make sure to cover all these cases and list the conditions that you will find these values.

To show that the Chebyshev polynomials \(T_n(x)\) are orthogonal on \(x \in [-1, 1]\) with respect to the weight 

$$
\omega(x) = \frac{1}{\sqrt{1 - x^2}},
$$

we need to prove that 

$$
\int_{-1}^{1} \omega(x) T_n(x) T_m(x) \, dx = 
\begin{cases} 
a & \text{if } m = n \\ 
0 & \text{if } m \neq n 
\end{cases}
$$

where \(a\) is a finite constant.

### Step 1: Use the Chebyshev Polynomial Identity

From the identity 

$$
T_n(\cos \theta) = \cos(n \theta),
$$

we can rewrite \(T_n(x)\) in terms of \(\theta\):

$$
x = \cos \theta \quad \Rightarrow \quad T_n(x) = \cos(n \theta).
$$

### Step 2: Change of Variables

Substituting \(x = \cos \theta\), we find that the differential \(dx\) becomes:

$$
dx = -\sin \theta \, d\theta.
$$

Thus, the integral becomes:

$$
\int_{-1}^{1} \omega(x) T_n(x) T_m(x) \, dx = \int_{-\pi/2}^{\pi/2} \frac{1}{\sqrt{1 - \cos^2 \theta}} \cos(n \theta) \cos(m \theta) (-\sin \theta) \, d\theta.
$$

Since \(\sqrt{1 - \cos^2 \theta} = \sin \theta\), we have:

$$
\int_{-\pi/2}^{\pi/2} \frac{1}{\sin \theta} \cos(n \theta) \cos(m \theta) (-\sin \theta) \, d\theta = \int_{-\pi/2}^{\pi/2} \cos(n \theta) \cos(m \theta) \, d\theta.
$$

### Step 3: Use the Product-to-Sum Identity

Using the product-to-sum identities, we find:

$$
\cos(n \theta) \cos(m \theta) = \frac{1}{2} \left( \cos((n-m)\theta) + \cos((n+m)\theta) \right).
$$

### Step 4: Integrate

Now, we can split the integral:

$$
\int_{-\pi/2}^{\pi/2} \cos(n \theta) \cos(m \theta) \, d\theta = \frac{1}{2} \left( \int_{-\pi/2}^{\pi/2} \cos((n-m)\theta) \, d\theta + \int_{-\pi/2}^{\pi/2} \cos((n+m)\theta) \, d\theta \right).
$$

### Step 5: Evaluate Each Integral

1. **For \(m \neq n\)**:

   The integral 

   $$
   \int_{-\pi/2}^{\pi/2} \cos(k \theta) \, d\theta = 0 \quad \text{for } k \neq 0.
   $$

   Therefore, both integrals are zero, leading to:

   $$
   \int_{-1}^{1} \omega(x) T_n(x) T_m(x) \, dx = 0 \quad \text{for } m \neq n.
   $$

2. **For \(m = n\)**:

   The integral 

   $$
   \int_{-\pi/2}^{\pi/2} \cos(0) \, d\theta = \int_{-\pi/2}^{\pi/2} 1 \, d\theta = \pi.
   $$

   Thus, we have:

   $$
   \int_{-1}^{1} \omega(x) T_n(x) T_n(x) \, dx = \frac{1}{2} \cdot \pi = \frac{\pi}{2}.
   $$

### Conclusion

From our calculations, we find that:

$$
\int_{-1}^{1} \omega(x) T_n(x) T_m(x) \, dx = 
\begin{cases} 
\frac{\pi}{2} & \text{if } m = n \\ 
0 & \text{if } m \neq n 
\end{cases}
$$

Thus, the Chebyshev polynomials are orthogonal with respect to the weight \(\omega(x) = \frac{1}{\sqrt{1 - x^2}}\), and the constant \(a\) is \(\frac{\pi}{2}\).

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-(x - 3)**2)

def relative_error(N, a, b):
    x_even = np.linspace(a, b, N)
    y_even = f(x_even)

    chebyshev_nodes = np.cos((2 * np.arange(N) + 1) * np.pi / (2 * N))
    x_chebyshev = (b - a) / 2 * (chebyshev_nodes + 1) + a
    y_chebyshev = f(x_chebyshev)

    x_dense = np.linspace(a, b, 400)

    deg_even = max(1, N // 3)
    coeffs_even = np.polyfit(x_even, y_even, deg_even)
    P_even = np.polyval(coeffs_even, x_dense)

    deg_chebyshev = max(1, N // 3)
    coeffs_chebyshev = np.polyfit(x_chebyshev, y_chebyshev, deg_chebyshev)
    P_chebyshev = np.polyval(coeffs_chebyshev, x_dense)

    err_even = f(x_dense) - P_even
    err_chebyshev = f(x_dense) - P_chebyshev

    r_even = np.linalg.norm(err_even) / np.linalg.norm(f(x_dense))
    r_chebyshev = np.linalg.norm(err_chebyshev) / np.linalg.norm(f(x_dense))

    return np.log10(r_even), np.log10(r_chebyshev)

N_values = np.arange(5, 31)
log_errors_even = []
log_errors_chebyshev = []

for N in N_values:
    log_r_even, log_r_chebyshev = relative_error(N, 0, 6)
    log_errors_even.append(log_r_even)
    log_errors_chebyshev.append(log_r_chebyshev)

plt.figure(figsize=(12, 8))
plt.plot(N_values, log_errors_even, label='Evenly Spaced Points', marker='o', color='red')
plt.plot(N_values, log_errors_chebyshev, label='Chebyshev Nodes', marker='o', color='blue')
plt.title('Log of Relative Error vs Number of Grid Points')
plt.xlabel('Number of Grid Points $N$')
plt.ylabel(r'$\log(r)$')
plt.grid()
plt.legend()
plt.xlim([5, 30])
plt.ylim([-5, 0])
plt.show()
````

# Finite-Difference Formulas for Equally Spaced Points

Let $ x_i, x_{i+1}, x_{i+2} $ be three equispaced points with spacing $ h $ such that:
- $ x_{i+1} = x_i + h $
- $ x_{i+2} = x_i + 2h $

## 1. First Derivative Approximations

### a. First Derivative at $ x_i $ (Forward Differences)

Assume the function $ f(x) $ can be expressed as a polynomial near $ x_i $:

$$
f(x) = f(x_i) + f'(x_i)(x - x_i) + \frac{f''(x_i)}{2}(x - x_i)^2 + \frac{f'''(x_i)}{6}(x - x_i)^3 + O((x - x_i)^4)
$$

Substituting $ x_{i+1} = x_i + h $:

$$
f(x_{i+1}) = f(x_i) + f'(x_i)h + \frac{f''(x_i)}{2}h^2 + \frac{f'''(x_i)}{6}h^3 + O(h^4)
$$

Now, we have:

$$
f'(x_i) = \frac{f(x_{i+1}) - f(x_i)}{h} - \frac{f''(x_i)}{2}h + O(h^2)
$$

Assume $ f'(x_i) = A + Bh + Ch^2 + O(h^3) $ for unknown coefficients $ A, B, C $. Thus:

$$
f'(x_i) \approx \frac{f(x_{i+1}) - f(x_i)}{h}
$$

Equating the coefficients of $ h $ leads to:

$$
A = f'(x_i) \quad \text{(exact value)}
$$
$$
B = -\frac{1}{2}f''(x_i) \quad \text{(leading to truncation error)}
$$
$$
C = O(h) \quad \text{(higher-order terms)}
$$

$$

$$

Using the same polynomial representation, we can express $ f(x_{i+2}) $ and $ f(x_{i}) $:

$$

$$
f(x_{i}) = f(x_{i+1}) - f'(x_{i+1})h + \frac{f''(x_{i+1})}{2}h^2 - O(h^3)

Subtracting these gives:

Rearranging gives:

Equating coefficients yields:

# Consistent 2nd Order First Derivative Matrix $ D $

Let $ y_i = f(x_i) $ be an equispaced set of points with spacing $ h $. We want to describe the structure of a consistent 2nd order first derivative matrix $ D $ such that

$$
f'(\mathbf{x}) \approx D\mathbf{y}
$$

returns the consistent first derivative at every point, including the endpoints.

## Structure of the Matrix $ D $

Assuming we have $ n $ equispaced points $ x_1, x_2, \ldots, x_n $ with spacing $ h $, the matrix $ D $ will be an $ n \times n $ matrix with the following structure:

1. **First Row** (Forward Difference at $ x_1 $):
   The first row approximates the first derivative at $ x_1 $ using the forward difference method:
   $$
   D_{1} = \begin{pmatrix}
   -3 & 4 & -1 & 0 & \cdots & 0
   \end{pmatrix}
   $$

   This corresponds to:
   $$
   f'(x_1) \approx \frac{1}{2h} \left( -3y_1 + 4y_2 - y_3 \right).
   $$

2. **Interior Rows** ($ i = 2, 3, \ldots, n-1 $):
   For the interior points, we use the centered difference formula:
   $$
   D_{i} = \begin{pmatrix}
   0 & \cdots & -1 & 0 & 1 & 0 & \cdots & 0
   \end{pmatrix}
   $$
   where the $ -1 $ is at position $ i-1 $ and $ 1 $ is at position $ i+1 $.

   This gives:
   $$
   f'(x_i) \approx \frac{1}{2h} \left( y_{i+1} - y_{i-1} \right).
   $$

3. **Last Row** (Backward Difference at $ x_n $):
   The last row approximates the first derivative at $ x_n $ using the backward difference method:
   $$
   D_{n} = \begin{pmatrix}
   0 & 0 & 0 & \cdots & 1 & -4 & 3
   \end{pmatrix}
   $$

   This leads to:
   $$
   f'(x_n) \approx \frac{1}{2h} \left( y_{n-2} - 4y_{n-1} + 3y_n \right).
   $$

## Complete Matrix $ D $

Factoring out $ \frac{1}{2h} $, the complete matrix $ D $ looks like this:

$$
D = \frac{1}{2h}
\begin{pmatrix}
-3 & 4 & -1 & 0 & \cdots & 0 \\
0 & -1 & 0 & 1 & 0 & \cdots & 0 \\
0 & 0 & -1 & 0 & 1 & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \cdots & -1 & 0 & 1 \\
0 & 0 & 0 & \cdots & 1 & -4 & 3
\end{pmatrix}
$$

**(c)** [4] Using the function `fdcoeffV.py` described in class (and included with this homework), write a python function that takes in a numpy array of coordinates $x$ (not necessarily evenly spaced)  and returns a matrix $D$ that calculates the consistent 2nd-order 1st derivatives at each point using 3 nearest neighbors (be careful at the edges)

Given this matrix $D$, it is easy to test it for various functions and meshes

````python
from  fdcoeffV import fdcoeffV
import numpy as np
import numpy

def D1(x):
    """
    Write a good doc-string
    
    x: ndarray of coordinates
    D: ndarray
        Matrix (2-D ndarray ) to calculate consistent 2nd order first derivative
    """
    # YOUR CODE HERE
    n = len(x)
    D = np.zeros((n, n))

    for i in range(n):
        if i == 0:
            ni = np.array([0, 1, 2])
        elif i == n - 1:
            ni = np.array([-3, -2, -1])
        else:
            ni = np.argsort(np.abs(x - x[i]))[:3]

        neighbors = x[ni]
        c = fdcoeffV(1, x[i], neighbors)
        D[i, ni] = c

    return D
````

````python
# Let's test this using f(x) = sin(pi*x) on a uniform mesh
N = 200
x = numpy.linspace(-1, 1, N)
dx = x[1]-x[0]

pi = numpy.pi
y = lambda x: numpy.sin(pi*x)
y_prime = lambda x: pi*numpy.cos(pi*x)

print("Uniform mesh: ",end='')
numpy.testing.assert_allclose(numpy.dot(D1(x), y(x)), y_prime(x), rtol=1e-2)
print("Success!")

# test on a randomized mesh
x_rand = x + dx/3.*numpy.random.rand(N)
print("Random mesh: ",end ='')
numpy.testing.assert_allclose(numpy.dot(D1(x_rand), y(x_rand)), y_prime(x_rand), rtol=1e-2)
print("Success!")

````

**(d)** [6] make a beautiful convergence plot showing how the relative error in your finite element approximation

$$ r = \frac{||f'(\mathbf{x}) - D\mathbf{y}||}{||f'(\mathbf{x})||}$$

 scales with mesh spacing $\Delta x$ for $f(x) = \sin(\pi x)$ on the interval $x\in[-1,1]$.  What is the expected rate of convergence?  Compare with the best fit straight line (in log space) to your error estimates.

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(np.pi * x)

def f_prime(x):
    return np.pi * np.cos(np.pi * x)

def relative_error(D, y, x):
    true_derivative = f_prime(x)
    return np.linalg.norm(true_derivative - D @ y) / np.linalg.norm(true_derivative)

mesh_sizes = [2**k for k in range(3, 11)]
errors = []

for mesh_size in mesh_sizes:
    x = np.linspace(-1, 1, mesh_size)
    y = f(x)
    D = D1(x)
    
    r = relative_error(D, y, x)
    errors.append(r)

deltas = [2 / (mesh_size - 1) for mesh_size in mesh_sizes]

plt.figure(figsize=(10, 6))
plt.loglog(deltas, errors, marker='o', label='Relative Error')
plt.xlabel(r'Mesh Spacing $\Delta x$', fontsize=14)
plt.ylabel(r'Relative Error $r$', fontsize=14)
plt.title('Convergence Plot of Finite Difference Approximation', fontsize=16)

coefficients = np.polyfit(np.log(deltas), np.log(errors), 1)
fit_line = np.exp(coefficients[1]) * np.array(deltas) ** coefficients[0]
plt.loglog(deltas, fit_line, '--', label=f'Best Fit Line: slope = {coefficients[0]:.2f}', color='red')

plt.legend()
plt.grid(True)
plt.show()
````

Rate of Convergence : 2.05
